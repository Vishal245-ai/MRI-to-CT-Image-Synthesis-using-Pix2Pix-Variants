```python
# ======================================
# Final Pix2Pix (MRI -> CT) Training Script
# With SSIM, PSNR, MAE, MSE evaluation
# ======================================
import os
import json
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
from tqdm import tqdm
from skimage.metrics import peak_signal_noise_ratio, structural_similarity, mean_squared_error
from sklearn.metrics import mean_absolute_error
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------
# Config
# -------------------------
class Config:
    data_root = "/kaggle/input/synth2023"
    batch_size = 64
    num_epochs = 75
    lr = 2e-4
    beta1 = 0.5
    image_size = 256
    save_dir = "/kaggle/working/pix2pix_run"
    validate_every = 1
    sample_grid_n = 4
    val_csv = "val_history.csv"
    loss_csv = "loss_history.csv"

cfg = Config()
os.makedirs(cfg.save_dir, exist_ok=True)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
torch.backends.cudnn.benchmark = True

# -------------------------
# Dataset
# -------------------------
class MRICtDataset(Dataset):
    def __init__(self, split, split_json, mri_dir, ct_dir, image_size=256):
        with open(split_json, "r") as f:
            splits = json.load(f)
        key_map = {"train": "train_patients", "val": "val_patients", "test": "test_patients"}
        patient_ids = splits[key_map[split]]
        all_mri_files = sorted(os.listdir(mri_dir))
        self.files = [f for f in all_mri_files if any(pid in f for pid in patient_ids)]
        self.mri_dir, self.ct_dir = mri_dir, ct_dir
        self.transform = T.Compose([T.ToTensor()])  # already preprocessed

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fname = self.files[idx]
        mri_slice = np.load(os.path.join(self.mri_dir, fname)).astype(np.float32)
        ct_slice  = np.load(os.path.join(self.ct_dir, fname)).astype(np.float32)
        if mri_slice.ndim == 2: mri_slice = np.expand_dims(mri_slice, axis=-1)
        if ct_slice.ndim == 2:  ct_slice  = np.expand_dims(ct_slice, axis=-1)
        mri = self.transform(mri_slice)
        ct  = self.transform(ct_slice)
        return mri, ct

# -------------------------
# Load Data
# -------------------------
split_json = "/kaggle/input/synth2023/processed/patient_split.json"
mri_dir = "/kaggle/input/synth2023/processed/MRI_slices"
ct_dir  = "/kaggle/input/synth2023/processed/CT_slices"

train_dataset = MRICtDataset("train", split_json, mri_dir, ct_dir, cfg.image_size)
val_dataset   = MRICtDataset("val", split_json, mri_dir, ct_dir, cfg.image_size)
test_dataset  = MRICtDataset("test", split_json, mri_dir, ct_dir, cfg.image_size)

train_loader = DataLoader(train_dataset, batch_size=cfg.batch_size, shuffle=True,
                          num_workers=2, pin_memory=True, prefetch_factor=2)
val_loader   = DataLoader(val_dataset, batch_size=min(8, cfg.batch_size), shuffle=False,
                          num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_dataset, batch_size=1, shuffle=False,
                          num_workers=2, pin_memory=True)

print(f"Train: {len(train_dataset)}, Val: {len(val_dataset)}, Test: {len(test_dataset)}")

# -------------------------
# Generator (U-Net)
# -------------------------
class UNetGenerator(nn.Module):
    def __init__(self, in_channels=1, out_channels=1, features=64):
        super().__init__()
        def down_block(in_ch, out_ch, apply_batchnorm=True):
            layers = [nn.Conv2d(in_ch, out_ch, 4, 2, 1, bias=False)]
            if apply_batchnorm: layers.append(nn.BatchNorm2d(out_ch))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return nn.Sequential(*layers)
        def up_block(in_ch, out_ch, apply_dropout=False):
            layers = [nn.ConvTranspose2d(in_ch, out_ch, 4, 2, 1, bias=False),
                      nn.BatchNorm2d(out_ch),
                      nn.ReLU(inplace=True)]
            if apply_dropout: layers.append(nn.Dropout(0.5))
            return nn.Sequential(*layers)
        self.down1 = down_block(in_channels, features, apply_batchnorm=False)
        self.down2 = down_block(features, features*2)
        self.down3 = down_block(features*2, features*4)
        self.down4 = down_block(features*4, features*8)
        self.down5 = down_block(features*8, features*8)
        self.down6 = down_block(features*8, features*8)
        self.down7 = down_block(features*8, features*8)
        self.down8 = down_block(features*8, features*8, apply_batchnorm=False)
        self.up1 = up_block(features*8, features*8, apply_dropout=True)
        self.up2 = up_block(features*16, features*8, apply_dropout=True)
        self.up3 = up_block(features*16, features*8, apply_dropout=True)
        self.up4 = up_block(features*16, features*8)
        self.up5 = up_block(features*16, features*4)
        self.up6 = up_block(features*8, features*2)
        self.up7 = up_block(features*4, features)
        self.final = nn.Sequential(nn.ConvTranspose2d(features*2, out_channels, 4, 2, 1), nn.Tanh())

    def forward(self, x):
        d1 = self.down1(x); d2 = self.down2(d1); d3 = self.down3(d2)
        d4 = self.down4(d3); d5 = self.down5(d4); d6 = self.down6(d5)
        d7 = self.down7(d6); bottleneck = self.down8(d7)
        u1 = self.up1(bottleneck)
        u2 = self.up2(torch.cat([u1, d7], 1))
        u3 = self.up3(torch.cat([u2, d6], 1))
        u4 = self.up4(torch.cat([u3, d5], 1))
        u5 = self.up5(torch.cat([u4, d4], 1))
        u6 = self.up6(torch.cat([u5, d3], 1))
        u7 = self.up7(torch.cat([u6, d2], 1))
        return self.final(torch.cat([u7, d1], 1))

# -------------------------
# Discriminator (PatchGAN)
# -------------------------
class PatchDiscriminator(nn.Module):
    def __init__(self, in_channels=2, features=64):
        super().__init__()
        def block(in_ch, out_ch, stride=2, normalize=True):
            layers = [nn.Conv2d(in_ch, out_ch, 4, stride, 1, bias=False)]
            if normalize: layers.append(nn.BatchNorm2d(out_ch))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return nn.Sequential(*layers)
        self.model = nn.Sequential(
            block(in_channels, features, normalize=False),
            block(features, features*2),
            block(features*2, features*4),
            block(features*4, features*8, stride=1),
            nn.Conv2d(features*8, 1, 4, 1, 1)
        )
    def forward(self, x, y):
        return self.model(torch.cat([x, y], dim=1))

# -------------------------
# Training Setup
# -------------------------
netG, netD = UNetGenerator().to(device), PatchDiscriminator().to(device)
criterion_GAN, criterion_L1 = nn.BCEWithLogitsLoss(), nn.L1Loss()
optimizer_G = optim.Adam(netG.parameters(), lr=cfg.lr, betas=(cfg.beta1, 0.999))
optimizer_D = optim.Adam(netD.parameters(), lr=cfg.lr, betas=(cfg.beta1, 0.999))
scaler = torch.amp.GradScaler("cuda")

resume_path = os.path.join(cfg.save_dir, "latest_checkpoint.pth")
start_epoch = 0

# Reload loss history if exists
loss_csv = os.path.join(cfg.save_dir, cfg.loss_csv)
if os.path.exists(loss_csv):
    loss_df = pd.read_csv(loss_csv)
    train_losses_G = loss_df["loss_G"].tolist()
    train_losses_D = loss_df["loss_D"].tolist()
    print(f"Loaded {len(train_losses_G)} epochs of loss history")
else:
    train_losses_G, train_losses_D = [], []

# Reload validation history if exists
val_csv = os.path.join(cfg.save_dir, cfg.val_csv)
if os.path.exists(val_csv):
    val_history = pd.read_csv(val_csv).to_dict("records")
    print(f" Loaded {len(val_history)} epochs of validation history")
else:
    val_history = []

# Resume from checkpoint if available
if os.path.exists(resume_path):
    checkpoint = torch.load(resume_path, map_location=device)
    netG.load_state_dict(checkpoint["netG"]); netD.load_state_dict(checkpoint["netD"])
    optimizer_G.load_state_dict(checkpoint["optG"]); optimizer_D.load_state_dict(checkpoint["optD"])
    if "scaler" in checkpoint: scaler.load_state_dict(checkpoint["scaler"])
    start_epoch = checkpoint.get("epoch", 0) + 1
    print(f"Resumed from epoch {start_epoch}")

# -------------------------
# Evaluation
# -------------------------
def evaluate_model(netG, loader):
    netG.eval()
    ssim_vals, psnr_vals, mae_vals, mse_vals = [], [], [], []
    with torch.no_grad():
        for mri, ct in loader:
            mri, ct = mri.to(device), ct.to(device)
            fake = netG(mri)
            ct_np, fake_np = ct.squeeze(1).cpu().numpy(), fake.squeeze(1).cpu().numpy()
            for gt, pred in zip(ct_np, fake_np):
                ssim_vals.append(structural_similarity(gt, pred, data_range=2.0))
                psnr_vals.append(peak_signal_noise_ratio(gt, pred, data_range=2.0))
                mae_vals.append(mean_absolute_error(gt, pred))
                mse_vals.append(mean_squared_error(gt, pred))
    return float(np.mean(ssim_vals)), float(np.mean(psnr_vals)), float(np.mean(mae_vals)), float(np.mean(mse_vals))

# -------------------------
# Save Sample Grids (with labels)
# -------------------------
def save_samples(netG, loader, epoch, save_dir, n=4):
    netG.eval()
    mri, ct = next(iter(loader))
    mri, ct = mri.to(device), ct.to(device)
    with torch.no_grad():
        fake = netG(mri[:n])

    mri_np  = mri[:n].squeeze(1).cpu().numpy()
    fake_np = fake.squeeze(1).cpu().numpy()
    ct_np   = ct[:n].squeeze(1).cpu().numpy()

    fig, axes = plt.subplots(n, 3, figsize=(9, 3*n))
    for i in range(n):
        axes[i,0].imshow(mri_np[i], cmap="gray");  axes[i,0].set_title("MRI");      axes[i,0].axis("off")
        axes[i,1].imshow(fake_np[i], cmap="gray"); axes[i,1].set_title("Fake CT");  axes[i,1].axis("off")
        axes[i,2].imshow(ct_np[i], cmap="gray");   axes[i,2].set_title("Real CT");  axes[i,2].axis("off")

    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"samples_epoch{epoch}.png"))
    plt.close(fig)
    netG.train()

# -------------------------
# Training Loop
# -------------------------
for epoch in range(start_epoch, cfg.num_epochs):
    netG.train(); netD.train()
    loop = tqdm(train_loader, leave=True)
    epoch_loss_G, epoch_loss_D = 0, 0

    for mri, ct in loop:
        mri, ct = mri.to(device), ct.to(device)

        # --- Train G ---
        optimizer_G.zero_grad()
        with torch.amp.autocast("cuda"):
            fake_ct = netG(mri)
            pred_fake = netD(mri, fake_ct)
            valid = torch.ones_like(pred_fake, device=device)
            loss_G = 0*criterion_GAN(pred_fake, valid) + criterion_L1(fake_ct, ct) * 030.0
        scaler.scale(loss_G).backward(); scaler.step(optimizer_G); scaler.update()

        # --- Train D ---
        optimizer_D.zero_grad()
        with torch.amp.autocast("cuda"):
            loss_real = criterion_GAN(netD(mri, ct), torch.ones_like(pred_fake, device=device))
            loss_fake = criterion_GAN(netD(mri, fake_ct.detach()), torch.zeros_like(pred_fake, device=device))
            loss_D = 1*(loss_real + loss_fake)
        scaler.scale(loss_D).backward(); scaler.step(optimizer_D); scaler.update()

        epoch_loss_G += loss_G.item(); epoch_loss_D += loss_D.item()
        loop.set_description(f"Epoch [{epoch+1}/{cfg.num_epochs}]")
        loop.set_postfix(loss_G=loss_G.item(), loss_D=loss_D.item())

    avg_loss_G = epoch_loss_G / len(train_loader)
    avg_loss_D = epoch_loss_D / len(train_loader)
    train_losses_G.append(avg_loss_G); train_losses_D.append(avg_loss_D)

    # Save/append losses to CSV
    pd.DataFrame([{"epoch": epoch+1, "loss_G": avg_loss_G, "loss_D": avg_loss_D}]) \
        .to_csv(loss_csv, mode="a", header=not os.path.exists(loss_csv), index=False)

    # --- Validation ---
    if (epoch+1) % cfg.validate_every == 0:
        ssim, psnr, mae, mse = evaluate_model(netG, val_loader)
        val_history.append({"epoch": epoch+1, "SSIM": ssim, "PSNR": psnr, "MAE": mae, "MSE": mse})
        pd.DataFrame(val_history).to_csv(val_csv, index=False)

        print(f"\nEpoch {epoch+1}: G {avg_loss_G:.4f}, "
              f"D {avg_loss_D:.4f}, "
              f"SSIM {ssim:.4f}, PSNR {psnr:.2f}, MAE {mae:.4f}, MSE {mse:.6f}")

        save_samples(netG, val_loader, epoch+1, cfg.save_dir, cfg.sample_grid_n)
        torch.save(netG.state_dict(), os.path.join(cfg.save_dir, f"netG_epoch{epoch+1}.pth"))
        torch.save({"epoch": epoch,
                    "netG": netG.state_dict(), "netD": netD.state_dict(),
                    "optG": optimizer_G.state_dict(), "optD": optimizer_D.state_dict(),
                    "scaler": scaler.state_dict()}, resume_path)

# -------------------------
# Final Plots
# -------------------------
plt.figure(figsize=(10,4))
plt.plot(train_losses_G, label="Generator Loss")
plt.plot(train_losses_D, label="Discriminator Loss")
plt.legend(); plt.title("Training Losses"); plt.xlabel("Epoch"); plt.ylabel("Loss")
plt.savefig(os.path.join(cfg.save_dir, "loss_curves.png"))

if len(val_history) > 0:
    pd.DataFrame(val_history).plot(x="epoch", y=["SSIM","MAE"], marker="o", figsize=(10,4))
    plt.title("Validation Metrics"); plt.xlabel("Epoch")
    plt.savefig(os.path.join(cfg.save_dir, "val_metrics.png"))

print(" Training complete. Plots, CSVs, and sample images saved.")
```
