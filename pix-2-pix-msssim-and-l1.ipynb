```python
# =========================================================
# Pix2Pix with MS-SSIM + L1 + Best Model Saving
# =========================================================
import os
import json
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.utils as vutils
from tqdm import tqdm
from skimage.metrics import peak_signal_noise_ratio
import pandas as pd
import matplotlib.pyplot as plt
import random
from pytorch_msssim import ms_ssim   # NEW

# -------------------------
# Config
# -------------------------
class Config:
    data_root = "/kaggle/input/synth2023"
    batch_size = 64
    num_epochs = 80
    lr = 2e-4
    beta1 = 0.5
    image_size = 256
    save_dir = "/kaggle/working/models"
    validate_every = 5
    sample_grid_n = 4
    max_val_batches = 50
    hu_min, hu_max = -1000, 2000

cfg = Config()
os.makedirs(cfg.save_dir, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
torch.backends.cudnn.benchmark = True

# -------------------------
# Loss Weights
# -------------------------
lambda_GAN = 1.0
lambda_L1 = 100.0
lambda_SSIM = 8.0

# -------------------------
# Reproducibility
# -------------------------
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

def worker_init_fn(worker_id):
    np.random.seed(seed + worker_id)
    random.seed(seed + worker_id)

g = torch.Generator().manual_seed(seed)

# -------------------------
# HU helper
# -------------------------
def to_hu(x):
    """Convert network output in [-1, 1] to CT HU."""
    return 0.5 * (x + 1.0) * (cfg.hu_max - cfg.hu_min) + cfg.hu_min

# -------------------------
# Dataset
# -------------------------
class MRICtDataset(Dataset):
    def __init__(self, split, split_json, mri_dir, ct_dir, image_size=256):
        with open(split_json, "r") as f:
            splits = json.load(f)
        self.files = splits["files"][split]
        self.mri_dir, self.ct_dir = mri_dir, ct_dir
        self.image_size = image_size
    def __len__(self): return len(self.files)
    def __getitem__(self, idx):
        fname = self.files[idx]
        mri_slice = np.load(os.path.join(self.mri_dir, fname)).astype(np.float32)
        ct_slice  = np.load(os.path.join(self.ct_dir, fname)).astype(np.float32)
        if mri_slice.ndim == 2: mri_slice = np.expand_dims(mri_slice, 0)
        if ct_slice.ndim == 2: ct_slice = np.expand_dims(ct_slice, 0)
        return torch.from_numpy(mri_slice), torch.from_numpy(ct_slice)

# -------------------------
# Data Loaders
# -------------------------
split_json = "/kaggle/input/synth2023/processed/patient_split.json"
mri_dir = "/kaggle/input/synth2023/processed/MRI_slices"
ct_dir  = "/kaggle/input/synth2023/processed/CT_slices"

train_dataset = MRICtDataset("train", split_json, mri_dir, ct_dir, cfg.image_size)
val_dataset   = MRICtDataset("val", split_json, mri_dir, ct_dir, cfg.image_size)

train_loader = DataLoader(train_dataset, batch_size=cfg.batch_size, shuffle=True,
    num_workers=2, pin_memory=(device.type=="cuda"),
    worker_init_fn=worker_init_fn, generator=g)
val_loader   = DataLoader(val_dataset, batch_size=min(8, cfg.batch_size), shuffle=False,
    num_workers=0, pin_memory=(device.type=="cuda"),
    worker_init_fn=worker_init_fn, generator=g)

print(f"Train: {len(train_dataset)}, Val: {len(val_dataset)}")

# -------------------------
# Generator & Discriminator
# -------------------------
class UNetGenerator(nn.Module):
    def __init__(self, in_channels=1, out_channels=1, features=64):
        super().__init__()
        def down_block(in_ch, out_ch, bn=True):
            layers = [nn.Conv2d(in_ch, out_ch, 4, 2, 1, bias=False)]
            if bn: layers.append(nn.BatchNorm2d(out_ch))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return nn.Sequential(*layers)
        def up_block(in_ch, out_ch, dropout=False):
            layers = [nn.ConvTranspose2d(in_ch, out_ch, 4, 2, 1, bias=False),
                      nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True)]
            if dropout: layers.append(nn.Dropout(0.5))
            return nn.Sequential(*layers)
        self.down1 = down_block(in_channels, features, bn=False)
        self.down2 = down_block(features, features*2)
        self.down3 = down_block(features*2, features*4)
        self.down4 = down_block(features*4, features*8)
        self.down5 = down_block(features*8, features*8)
        self.down6 = down_block(features*8, features*8)
        self.down7 = down_block(features*8, features*8)
        self.down8 = down_block(features*8, features*8, bn=False)
        self.up1 = up_block(features*8, features*8, dropout=True)
        self.up2 = up_block(features*16, features*8, dropout=True)
        self.up3 = up_block(features*16, features*8, dropout=True)
        self.up4 = up_block(features*16, features*8)
        self.up5 = up_block(features*16, features*4)
        self.up6 = up_block(features*8, features*2)
        self.up7 = up_block(features*4, features)
        self.final = nn.Sequential(nn.ConvTranspose2d(features*2, out_channels, 4, 2, 1), nn.Tanh())
    def forward(self, x):
        d1=self.down1(x); d2=self.down2(d1); d3=self.down3(d2)
        d4=self.down4(d3); d5=self.down5(d4); d6=self.down6(d5)
        d7=self.down7(d6); bottleneck=self.down8(d7)
        u1=self.up1(bottleneck)
        u2=self.up2(torch.cat([u1,d7],1))
        u3=self.up3(torch.cat([u2,d6],1))
        u4=self.up4(torch.cat([u3,d5],1))
        u5=self.up5(torch.cat([u4,d4],1))
        u6=self.up6(torch.cat([u5,d3],1))
        u7=self.up7(torch.cat([u6,d2],1))
        return self.final(torch.cat([u7,d1],1))

class PatchDiscriminator(nn.Module):
    def __init__(self, in_channels=2, features=64):
        super().__init__()
        def block(in_ch, out_ch, stride=2, normalize=True):
            layers=[nn.Conv2d(in_ch,out_ch,4,stride,1,bias=False)]
            if normalize: layers.append(nn.BatchNorm2d(out_ch))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return nn.Sequential(*layers)
        self.model=nn.Sequential(
            block(in_channels,features,normalize=False),
            block(features,features*2),
            block(features*2,features*4),
            block(features*4,features*8,stride=1),
            nn.Conv2d(features*8,1,4,1,1)
        )
    def forward(self,x,y): return self.model(torch.cat([x,y],dim=1))

# -------------------------
# Setup
# -------------------------
netG = UNetGenerator().to(device)
netD = PatchDiscriminator().to(device)

criterion_GAN = nn.BCEWithLogitsLoss()
criterion_L1 = nn.L1Loss()

optimizer_G = optim.Adam(netG.parameters(), lr=cfg.lr, betas=(cfg.beta1,0.999))
optimizer_D = optim.Adam(netD.parameters(), lr=cfg.lr, betas=(cfg.beta1,0.999))

scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer_G, mode="min", factor=0.5,
                                                patience=3, verbose=True)

scaler = torch.amp.GradScaler("cuda")

# -------------------------
# Evaluation
# -------------------------
def evaluate_model(netG, loader, max_batches=None, save_samples=False, epoch=0):
    netG.eval()
    mae_vals, psnr_vals, ssim_vals = [], [], []
    sample_batches = []

    with torch.no_grad():
        for i,(mri,ct) in enumerate(loader):
            mri,ct = mri.to(device), ct.to(device)
            fake_ct = netG(mri)
            ct_hu, fake_hu = to_hu(ct), to_hu(fake_ct)

            # Metrics
            mae_vals.append(torch.mean(torch.abs(ct_hu-fake_hu)).item())
            data_range=cfg.hu_max-cfg.hu_min
            ct_np, fake_np = ct_hu.squeeze(1).cpu().numpy(), fake_hu.squeeze(1).cpu().numpy()
            psnr_batch=[peak_signal_noise_ratio(gt,pr,data_range=data_range)
                        for gt,pr in zip(ct_np,fake_np)]
            psnr_vals.append(np.mean(psnr_batch))
            ct01=(ct_hu-cfg.hu_min)/data_range
            fake01=(fake_hu-cfg.hu_min)/data_range
            ssim_vals.append(ms_ssim(fake01,ct01,data_range=1.0,size_average=True).item())

            # Save first batch for visualization
            if save_samples and i == 0:
                sample_batches = (mri[:cfg.sample_grid_n],
                                  ct[:cfg.sample_grid_n],
                                  fake_ct[:cfg.sample_grid_n])

            if max_batches and (i+1>=max_batches): break

    # Save visualization grid
    if save_samples and sample_batches:
        mri_s, ct_s, fake_s = sample_batches
        grid = torch.cat([mri_s, ct_s, fake_s], dim=0)  # stack rows
        vutils.save_image(grid, 
            os.path.join(cfg.save_dir, f"val_samples_epoch{epoch}.png"),
            nrow=cfg.sample_grid_n, normalize=True, scale_each=True)
    
    return (np.mean(mae_vals),np.mean(psnr_vals),np.mean(ssim_vals))

# -------------------------
# Training Loop
# -------------------------
G_losses,D_losses,val_history=[],[],[]
best_ssim=-1; best_mae=float("inf")

for epoch in range(cfg.num_epochs):
    netG.train(); netD.train()
    loop=tqdm(train_loader,leave=False)
    for mri,ct in loop:
        mri,ct = mri.to(device), ct.to(device)
        # Generator
        optimizer_G.zero_grad()
        with torch.amp.autocast(device_type="cuda",enabled=(device.type=="cuda")):
            fake_ct=netG(mri)
            pred_fake=netD(mri,fake_ct)
            loss_GAN=criterion_GAN(pred_fake,torch.ones_like(pred_fake,device=device))
            loss_L1=criterion_L1(fake_ct,ct)
            ct_hu,fake_hu=to_hu(ct),to_hu(fake_ct)
            ct01=(ct_hu-cfg.hu_min)/(cfg.hu_max-cfg.hu_min)
            fake01=(fake_hu-cfg.hu_min)/(cfg.hu_max-cfg.hu_min)
            loss_SSIM=1.0-ms_ssim(fake01,ct01,data_range=1.0,size_average=True)
            loss_G=lambda_GAN*loss_GAN + lambda_L1*loss_L1 + lambda_SSIM*loss_SSIM
        scaler.scale(loss_G).backward()
        scaler.step(optimizer_G); scaler.update()
        # Discriminator
        optimizer_D.zero_grad()
        with torch.amp.autocast(device_type="cuda",enabled=(device.type=="cuda")):
            pred_real=netD(mri,ct)
            loss_real=criterion_GAN(pred_real,torch.ones_like(pred_real,device=device))
            pred_fake_det=netD(mri,fake_ct.detach())
            loss_fake=criterion_GAN(pred_fake_det,torch.zeros_like(pred_fake_det,device=device))
            loss_D=0.5*(loss_real+loss_fake)
        scaler.scale(loss_D).backward()
        scaler.step(optimizer_D); scaler.update()
        loop.set_description(f"Epoch [{epoch+1}/{cfg.num_epochs}]")
        loop.set_postfix(loss_G=loss_G.item(),loss_D=loss_D.item())
    G_losses.append(loss_G.item()); D_losses.append(loss_D.item())
    # Validation
    if (epoch+1)%cfg.validate_every==0:
       mae,psnr,ssim = evaluate_model(netG, val_loader, cfg.max_val_batches,save_samples=True, epoch=epoch+1)
       print(f"Epoch {epoch+1} Val -> MAE: {mae:.2f} HU, PSNR: {psnr:.2f}, MS-SSIM: {ssim:.4f}")
       val_history.append({"epoch":epoch+1,"MAE":mae,"PSNR":psnr,"SSIM":ssim})
       scheduler.step(mae)  # step scheduler with MAE
       # Save best model
       if (ssim>best_ssim) or (ssim==best_ssim and mae<best_mae):
           best_ssim, best_mae = ssim, mae
           torch.save(netG.state_dict(), os.path.join(cfg.save_dir,"best_model.pth"))
           print(f"New best model saved at epoch {epoch+1} (MAE={mae:.2f}, SSIM={ssim:.4f})")

# -------------------------
# Save Results
# -------------------------
pd.DataFrame(val_history).to_csv(os.path.join(cfg.save_dir,"val_history.csv"),index=False)
plt.figure(); plt.plot(G_losses,label="G"); plt.plot(D_losses,label="D"); plt.legend()
plt.savefig(os.path.join(cfg.save_dir,"loss_curve.png"))
print("Training complete. Best model at:", os.path.join(cfg.save_dir,"best_model.pth"))
print(f"Best Val -> MAE: {best_mae:.2f} HU, MS-SSIM: {best_ssim:.4f}")
```
